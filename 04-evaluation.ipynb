```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df =pd.DataFrame({
    "model": ["Majority","25-NN", "DT", "RF", "XGBoost"],
    "accuracy": [0.209,0.232, 0.235,0.30, 0.31]
})
```

```python
# Balkendiagramm erstellen
plt.figure(figsize=(10, 6))
plt.bar(df['model'], df['accuracy'], color='skyblue')
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.title('Accuracy Comparison')
plt.show()
```
